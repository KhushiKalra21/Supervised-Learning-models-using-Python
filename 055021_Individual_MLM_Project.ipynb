```python
# DATA PREPROCESSING

# 1.Treatment of Missing Data
# Numerical columns: Impute missing values with the mean
kk21_numerical_columns = kk21_sample.select_dtypes(include=np.number).columns
kk21_imputer_mean = SimpleImputer(strategy="mean")
kk21_sample[kk21_numerical_columns] = kk21_imputer_mean.fit_transform(kk21_sample[kk21_numerical_columns])

# Categorical columns: Impute missing values with the most frequent value
kk21_categorical_columns = kk21_sample.select_dtypes(include=["object"]).columns
kk21_imputer_mode = SimpleImputer(strategy="most_frequent")
kk21_sample[kk21_categorical_columns] = kk21_imputer_mode.fit_transform(kk21_sample[kk21_categorical_columns])

print("Missing data treated successfully!")

# 2.Numerical Encoding of Non-Numeric Categorical Data
# Label Encoding for categorical variables
kk21_label_encoder = LabelEncoder()
for kk21_col in kk21_categorical_columns:
    kk21_sample[kk21_col] = kk21_label_encoder.fit_transform(kk21_sample[kk21_col])

print("Categorical data encoded successfully!")

# 3.Rescaling & Transformation of Numeric Non-Categorical Data (Treatment of Outliers)
# Outlier treatment using RobustScaler
# RobustScaler is robust to outliers by removing the median and scaling using the IQR
kk21_scaler = RobustScaler()
kk21_sample[kk21_numerical_columns] = kk21_scaler.fit_transform(kk21_sample[kk21_numerical_columns])

print("Numeric data rescaled and transformed successfully!")

# Display processed data
print("Processed dataset preview:")
print(kk21_sample.head())
```

# Inferential Statistics

```python
# INFERENTIAL STATISTICS

# 1.Test of Homogeneity for Categorical Variable

def test_homogeneity_categorical(df, col1, col2):
    """
    Perform Chi-Square Test of Homogeneity between two categorical variables.
    """
    contingency_table = pd.crosstab(df[col1], df[col2])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"Chi-Square Test for {col1} vs {col2}")
    print(f"Chi-Square Statistic: {chi2:.4f}, p-value: {p:.4f}, Degrees of Freedom: {dof}")
    print("Expected Frequencies:\n", expected)
    if p < 0.05:
        print("Reject Null Hypothesis: Significant relationship exists.\n")
    else:
        print("Fail to Reject Null Hypothesis: No significant relationship.\n")

# 2. Test of Normality for Non-Categorical Variable

def test_normality(df, col):
    """
    Perform multiple tests of normality on a non-categorical variable.
    """
    print(f"Normality Tests for {col}")
    # Shapiro-Wilk Test
    stat, p = stats.shapiro(df[col])
    print(f"Shapiro-Wilk Test: Statistic={stat:.4f}, p-value={p:.4f}")

    # Kolmogorov-Smirnov Test
    stat, p = stats.kstest(df[col], 'norm', args=(df[col].mean(), df[col].std()))
    print(f"Kolmogorov-Smirnov Test: Statistic={stat:.4f}, p-value={p:.4f}")

    # Anderson-Darling Test
    result = normal_ad(df[col])
    print(f"Anderson-Darling Test: Statistic={result[0]:.4f}, p-value={result[1]:.4f}")

    # Jarque-Bera Test
    stat, p = stats.jarque_bera(df[col])
    print(f"Jarque-Bera Test: Statistic={stat:.4f}, p-value={p:.4f}")

    print("\n")

# 3. Test of Correlation for Non-Categorical Variable

def test_correlation(df, col1, col2):
    """
    Perform a t-test of correlation between two non-categorical variables.
    """
    stat, p = stats.pearsonr(df[col1], df[col2])
    print(f"T-Test for Correlation between {col1} and {col2}")
    print(f"Pearson Correlation Coefficient: {stat:.4f}, p-value: {p:.4f}")
    if p < 0.05:
        print("Significant Correlation Exists.\n")
    else:
        print("No Significant Correlation.\n")

# Run the Inferential Tests

# Test of Homogeneity (Chi-Square)
test_homogeneity_categorical(kk21_sample, 'Product_Category', 'Active_Member')

# Test of Normality
test_normality(kk21_sample, 'Yearly_Average_Salary')
test_normality(kk21_sample, 'Yearly_Average_Balance')

# Test of Correlation
test_correlation(kk21_sample, 'Age', 'Credit_Score')
test_correlation(kk21_sample, 'Yearly_Average_Balance', 'Tenure_Years')
```

### Unsupervised Machine Learning
●	Clustering: Cluster Characteristics
{Nature of each Cluster (Homogeneity) | Difference between Clusters (Heterogeneity)}




```python
# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(kk21_sample)

# K-Means clustering (adjust number of clusters as needed)
kmeans = KMeans(n_clusters=10, random_state=55021)  # Example: 3 clusters
kmeans.fit(X)

# Add cluster labels to the dataset
kk21_sample['Cluster'] = kmeans.labels_

# Cluster Characteristics: Homogeneity and Heterogeneity

# Homogeneity within each cluster
def calculate_homogeneity(X, labels):
    homogeneity = {}
    for label in np.unique(labels):
        cluster_points = X[labels == label]
        distances = pairwise_distances_argmin_min(cluster_points, cluster_points)
        homogeneity[label] = np.mean(distances[1])  # Mean distance within the cluster
    return homogeneity

homogeneity = calculate_homogeneity(X, kmeans.labels_)

# Heterogeneity between clusters
def calculate_heterogeneity(X, labels, centroids):
    heterogeneity = {}
    for label in np.unique(labels):
        cluster_points = X[labels == label]
        centroid = centroids[label]
        distances = np.linalg.norm(cluster_points - centroid, axis=1)
        heterogeneity[label] = np.mean(distances)  # Mean distance from centroid
    return heterogeneity

centroids = kmeans.cluster_centers_
heterogeneity = calculate_heterogeneity(X, kmeans.labels_, centroids)

# Output homogeneity and heterogeneity for each cluster
print("Homogeneity within each cluster:")
for label, value in homogeneity.items():
    print(f"Cluster {label}: {value}")

print("\nHeterogeneity between clusters:")
for label, value in heterogeneity.items():
    print(f"Cluster {label}: {value}")

# Optional: Visualize the clusters (2D)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('Cluster Visualization')
plt.show()
```
